# E-Commerce Sales Data Analysis

## Project Overview

This project analyzes e-commerce sales data using Python, Pandas, and Matplotlib to identify key business insights such as sales trends, regional performance, and product profitability.

The goal of this analysis is to help businesses understand their sales performance and identify opportunities for growth.

```python
import pandas as pd

df = pd.read_csv("ecommerce_sales_data.csv")

df.head()
```

## Data Overview

The dataset contains transactional e-commerce sales information including:

- Order Date
- Product Name
- Category
- Region
- Quantity
- Sales
- Profit

These variables allow us to analyze sales performance across different products, regions, and time periods.


```python
df.info()
```

```python
total_sales = df["Sales"].sum()

print("Total Sales:", total_sales)
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

print(category_sales)
```

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

print(top_products)
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
import matplotlib.pyplot as plt

category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.ylabel("Sales")
plt.show()
```

## Insights

- Electronics generated the highest sales.
- The West region shows strong purchasing activity.
- Printers and Tablets are among the top revenue-generating products.cts.

```python
import pandas as pd

df["Order Date"] = pd.to_datetime(df["Order Date"])

monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()

monthly_sales
```

```python
import matplotlib.pyplot as plt
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```

### Insight

Sales fluctuate across months, indicating seasonal demand patterns.
Some months show higher sales activity which could be related to promotions or increased market demand.

```python
df["Profit Margin"] = df["Profit"] / df["Sales"]
df["Profit Margin"].mean()
```

### Insight

The average profit margin provides insight into the overall profitability of sales.
Higher margins indicate better pricing strategies and efficient cost management.

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
region_sales
```

```python
region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.ylabel("Total Sales")
plt.show()
```

### Insight

Certain regions generate higher sales than others.  
This suggests stronger demand or better distribution channels in these regions.

## Final Insights

Key findings from the analysis:

- Some product categories generate significantly higher revenue.
- Sales vary across regions indicating different market demand levels.
- Monthly sales trends reveal potential seasonal patterns.
- Profit margin analysis helps evaluate overall business profitability.

These insights can help businesses improve pricing strategies, optimize inventory, and focus on high-performing regions and products.
